```python
n = 100
u = 1
s = 0.25
fig = go.Figure(
    data=[go.Mesh3d(
        x=(u + s * np.random.randn(n)),
        y=(u + s * np.random.randn(n)),
        z=(u + s * np.random.randn(n)),
        opacity=0.25,
    )]
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

def plot(x, y, z):
    fig = go.Figure(data=[go.Surface(
        x=x, y=y, z=z,
        colorscale='Viridis',
        opacity=0.9
    )])
    fig.update_layout(scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        aspectmode='cube'
    ))
    fig.show()

sample = np.linspace(-3, 3, 200)
x, y = np.meshgrid(sample, sample)

z = [
    np.sin(x**2 + y**2) * np.cos(2 * x) * np.sin(2 * y),
    np.sin(x**2 + y) * np.cos(10 * x) * np.sin(10 * y),
    np.sin(x**2 + y**2) + x**3 * np.sin(2 * y),
    x**2 + y**2,
]

plot(x, y, z[0])
```
